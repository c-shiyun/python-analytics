## Data cleaning

Q1: Why there is a need to apply data cleaning?

- Missing data: Na/NaN
- Duplicates data
- Incorrect data

Q2: What are the data cleaning methods?

- dropna
- replace
- ffill
- bfill

#### load and read data

```python
# Import libraries
import numpy as np
import pandas as pd

# Read csv
data = pd.read_csv("data/netflix_titles.csv") 
```

```python
# Display data
data
```

```python
# Check number of rows where particular columns of null values
data.isnull().sum()
```

From the list of show_id, it can be observed null values could be found under director, cast, country, data_added, rating and duration columns.

### NaN - apply dropna

```python
# 1. Data name no change and without applying inplace = True


# Remove null values
data = data.dropna()

# Display cleaned new data
data
```

```python
# 2. Added new data name and without applying inplace = True

# Remove null values
data_new = data.dropna()

# Display cleaned new data
data_new
```

```python
# 3. Data name no change and without applying inplace = True: changed original data


# Remove null values
data.dropna(inplace = True)

# Display cleaned new data
data
```

The number of columns have been reduced from 8807 to 5332 rows.

### Replace NaN with 0 (pandas)

```python
# Read csv
data = pd.read_csv("data/netflix_titles.csv") 

# Replace NaNs with 0s
data = data.fillna(0)

# Display replaced data
data
```

### Replace NaN with 0 (numpy)

```python
# Read csv
data = pd.read_csv("data/netflix_titles.csv") 

# Display data
data
```

```python
# Replace NaNs with 0s
data["cast"] = data["cast"].replace(np.nan, 0)

# Display data
data
```

It is shown that "cast" cloumn with NaNs has been replaced with 0s.

### Replace fillna with text

```python
# Read csv
data = pd.read_csv("data/netflix_titles.csv") 

# Display data
data
```

```python
# Replace NaNs with text
data["cast"].fillna("text", inplace = True)

# Display data
data
```

It is shown that "cast" cloumn with NaNs has been replaced with text.

## Forward fill - ffill: replace null values from previous row or previous column

### Replace ffill with text from previous row

```python
# Read csv
data = pd.read_csv("data/netflix_titles.csv") 

# Display data
data
```

```python
# Replace NaNs from previous row under same column
data["director"].fillna(method ='ffill', inplace = True)

# Display data
data
```

It has been observed that NaNs under "director" column has been replaced with the name from previous row under same column.

### Limit - replace ffill with text from previous row

```python
# Read csv
data = pd.read_csv("data/netflix_titles.csv") 

# Display data
data
```

```python
# Replace NaNs from previous row under same column
data["country"].fillna(method ='ffill', limit = 1, inplace = True)

# Display data
data
```

It has been observed that NaN under "country" column has been replaced with a name from previous row under same column.

## Backward fill - bfill: replace null values from next row or next column

### Replace bfill with text from next column 

```python
# Read csv
data = pd.read_csv("data/netflix_titles.csv") 

# Display data
data
```

```python
# Replace NaNs from next row under same column
data["director"].fillna(method ='bfill', inplace = True)

# Display data
data
```

### Limit - replace bfill with text from next row

```python
# Read csv
data = pd.read_csv("data/netflix_titles.csv") 

# Display data
data
```

```python
# Replace NaNs from previous row under same column
data["country"].fillna(method ='bfill', limit = 1, inplace = True)

# Display data
data
```
