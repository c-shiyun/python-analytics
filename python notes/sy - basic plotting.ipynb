## Plot with numpy

### Line plot

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Plot line (x,y)
xpoints = np.array([1, 2])
ypoints = np.array([3, 4])

plt.plot(xpoints, ypoints)

plt.title("Simple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Plot
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

### Bar plot

- Categorical data

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Plot bar
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([90, 75, 55, 43])

plt.bar(x,y, width = 0.5)

plt.title("Student Result")
plt.xlabel("Grade")
plt.ylabel("Mark")

# Plot
plt.show()
```

## Plot with matplotlib

### Line plot

```python
# Draw plot
%matplotlib inline

# Import library
import matplotlib.pyplot as plt

# Plot line (x,y)
# (0, 2), (1, 4), (2, 6), (3, 8), (4, 10)
plt.plot([2,4,6,8,10])

plt.title("Simple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Plot
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

### Bar plot

- Categorical data

```python
# Draw plot
%matplotlib inline

# Import library
import matplotlib.pyplot as plt

# Plot bar chart
plt.bar(['A', 'B', 'C', 'D'], [90, 75, 55, 43], width = 0.5)

plt.title("Student Result")
plt.xlabel("Grade")
plt.ylabel("Mark")

# Plot
plt.show()
```

### Scatter plot

```python
# Import library
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.title("Linear scatter plot")
plt.xlabel("X")
plt.ylabel("Y")

plt.scatter(x, y)
plt.show()
```

## Subplot

```python
# Import library
import matplotlib.pyplot as plt
import numpy as np

# Plot 1:
x = np.array([0, 1, 2])
y = np.array([3, 4, 5])

# Plot 1 rows 1 column first plot
plt.subplot(2, 1, 1)
plt.scatter(x,y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.title("Plot1")
plt.xlabel("X")
plt.ylabel("Y")

# Plot 2:
x = np.array([0, 1, 2])
y = np.array([10, 20, 30])

# Plot 2 rows 1 column second plot
plt.subplot(2, 1, 2)
plt.scatter(x,y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.title("Plot2")
plt.xlabel("X")
plt.ylabel("Y")

# Display subplot
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.6,
                    hspace=0.6)

plt.suptitle("Subplot of 2 rows 1 column")
plt.show()
```

```python
# Import library
import matplotlib.pyplot as plt
import numpy as np

# Plot 1:
x = np.array([0, 1, 2])
y = np.array([3, 4, 5])

# Plot 1 row 2 columns first plot
plt.subplot(1, 2, 1)
plt.bar(x,y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.title("Plot1")
plt.xlabel("X")
plt.ylabel("Y")

# Plot 2:
x = np.array([0, 1, 2])
y = np.array([10, 20, 30])

# Plot 1 row 2 columns second plot
plt.subplot(1, 2, 2)
plt.bar(x,y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.title("Plot2")
plt.xlabel("X")
plt.ylabel("Y")

# Display subplot
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.6,
                    hspace=0.6)

plt.suptitle("Subplot of 1 row 2 columns")
plt.show()
```
